```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("output.png")
img = img[:, :, :3]
```

```python
img_height, img_width, _ = img.shape
```

```python
img_biner = np.zeros(img.shape, dtype=np.uint8)
```

```python
for y in range(img_height):
    for x in range(img_width):
        red, green, blue = img[y, x]
        # Menggunakan rumus luminance
        gray = int(0.299 * red + 0.587 * green + 0.114 * blue)
        if gray < 128:
            biner = 0
        else:
            biner = 255
        img_biner[y, x] = (biner, biner, biner)
```

```python
plt.imshow(img_biner)
plt.title("Biner")
plt.show()
```
